```python
def afficher_matrice(matrice):
    for ligne in matrice:
        print(" ".join(f"{val:2}" for val in ligne))



def est_diagonale(matrice):
    return all(
        matrice[i][j] == 0
        for i in range(len(matrice))
        for j in range(len(matrice[i]))
        if i != j
    )


def est_symetrique(matrice):
    return all(
        matrice[i][j] == matrice[j][i]
        for i in range(len(matrice))
        for j in range(len(matrice[i]))
    )


def multiplication_vecteur(matrice, vecteur):
    return [
        sum(matrice[i][j] * vecteur[j] for j in range(len(vecteur)))
        for i in range(len(matrice))
    ]


def multiplication_matrices(A, B):
    if len(A[0]) != len(B):
        raise ValueError("Multiplication impossible, dimensions incompatibles.")
    return [
        [sum(A[i][k] * B[k][j] for k in range(len(B))) for j in range(len(B[0]))]
        for i in range(len(A))
    ]
```

```python
import cmath
import numpy as np
```

```python
def compute_w_matrix(N):
    W = np.zeros((N, N), dtype=complex)
    for n in range(N):
        for m in range(N):
            W[n, m] = cmath.exp(-2j * cmath.pi * n * m / N)
    return W


def format_matrix(W):
    def simplify(c):
        if abs(c.real) < 1e-10:
            if abs(c.imag - 1) < 1e-10:
                return " i"
            elif abs(c.imag + 1) < 1e-10:
                return "-i"
            elif abs(c.imag) < 1e-10:
                return 0
        elif abs(c.real - 1) < 1e-10:
            return " 1"
        elif abs(c.real + 1) < 1e-10:
            return "-1"
        return str(c)

    return [[simplify(W[n, m]) for m in range(W.shape[1])] for n in range(W.shape[0])]


# Exemple avec N = 4
N = 4
W_4 = compute_w_matrix(N)
afficher_matrice(format_matrix(W_4))
```

```python
from math import sin, pi 
import matplotlib.pyplot as plt
```

```python
def generate_signal(f, N):
    return [sin(2 * pi * f * k) for k in range(N)]

# Application pour f = 0.01 et N = 100
example_signal = generate_signal(0.01, 100)
print("\nSignal pour f = 0.01 et N = 100:")
print(example_signal)
```

```python
plt.plot(range(100), example_signal)
plt.xlabel("Échantillon k")
plt.ylabel("Amplitude")
plt.title("Signal courbe")
plt.show()
```

```python
def analyze_signal(signal):
    return max(signal), min(signal)


max_val, min_val = analyze_signal(example_signal)
print(f"Valeur maximale: {max_val}")
print(f"Valeur minimale: {min_val}")
```

```python
for freq in [0.01, 0.5, 0.075, 0.073]:
    test_signal = generate_signal(freq, 100)
    max_val, min_val = analyze_signal(test_signal)
    print(f"\nSignal pour f = {freq} et N = 100:")
    print(f"Valeur maximale: {max_val}")
    print(f"Valeur minimale: {min_val}")

    plt.plot(range(100), test_signal, label=f"f={freq}")

plt.xlabel("Échantillon k")
plt.ylabel("Amplitude")
plt.title("Signal sin pour différentes fréquences")
plt.legend()
plt.show()

print("""\nExplication :
Le sin de signal varie entre -1 et 1 en amplitude.

Pour f=0.01, la variation est lente, alors que pour f=0.5, on a une courbe de 0 car dans ce cas sin 2 pi f n  => sin 2 pi n, qui est une sinusoïde de période 1.

Des fréquences proches (comme 0.075 et 0.073) donnent des signaux similaires mais légèrement décalés.
"""
)
```
