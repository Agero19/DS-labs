```python
import numpy as np
import matplotlib.pyplot as plt
plt.close ('all')
a = 10
N = 2*a +1
n = np.linspace ( -a ,a , N )

def exponentielle ( n ) :
    alpha = 0.9
    exponentielle = np.where (n >=0 , np.power ( alpha , n ) ,0)
    return exponentielle


signal = exponentielle
plt.figure (1)
plt.stem (n , signal(n) )
axes = plt.gca ()
axes.set_ylim ([ -2 ,2])
axes.grid ( color = 'k' , linestyle = '-' , linewidth=0.2)
plt.title ( 'Signal Original S ( n ) ')
plt.stem(
    n,
    signal(n),
    linefmt="C0",
    markerfmt="C0o",
    basefmt="C0",
)
```

Le parametre a controle les valeurs sur l'axe des abcisses n = 2 * a + 1 definit le nombre total de points du signal et n prend des valeuts entre -a et a

```python


def gamma(n):
    return np.where(n >= 0, 1, 0)


# 1. S(n-2)
plt.figure(figsize=(10, 5))

plt.subplot(2, 3, 1)
plt.stem(n, signal(n - 2))
plt.ylim([-2, 2])
plt.title("S(n-2)")
plt.grid(True)

# 2. S(-n-5)
plt.subplot(2, 3, 2)
plt.stem(n, signal(-n - 5))
plt.ylim([-2, 2])
plt.title("S(-n-5)")
plt.grid(True)

# 3. Γ(n)
plt.subplot(2, 3, 3)
plt.stem(n, gamma(n))
plt.ylim([-2, 2])
plt.title("Γ(n)")
plt.grid(True)

# 4. Γ(n-2)
plt.subplot(2, 3, 4)
plt.stem(n, gamma(n - 2))
plt.ylim([-2, 2])
plt.title("Γ(n-2)")
plt.grid(True)

# 5. Γ(n) - Γ(n-5)
plt.subplot(2, 3, 5)
plt.stem(n, gamma(n) - gamma(n - 5))
plt.ylim([-2, 2])
plt.title("Γ(n) - Γ(n-5)")
plt.grid(True)

plt.tight_layout()
```

S(n-2)

Décalé de 2 unités vers la droite. L'exponentielle commence à n = 2 au lieu de n = 0, et décroît de 0,9^0 = 1 , à n = 2.

S(-n-5)

Le signal est non nul pour n ≤ -5, avec des valeurs de 0,9^(-n -5) . Par exemple :

À n = -5 : 0,9^0 = 1.
À n = -6 : 0,9^1 = 0, décroissant au fur et à mesure que nn devient plus négatif.

Γ(n)-Γ(n-5)
Une impulsion rectangulaire de n = 0 à n = 4. La soustraction annule les valeurs pour n ≥ 5, ce qui donne une impulsion à 5 échantillons.

```python
n = np.arange(0, 30, 1)  
s1 = np.sin((np.pi / 5) * n)
s2 = np.cos((3 * np.pi / 5) * n)
s3 = 0.5 * np.sin((np.pi / 2) * n)

plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.stem(n, s1)
plt.title(r'$s_1(n) = \sin(\frac{\pi}{5} n)$')
plt.grid()

plt.subplot(3, 1, 2)
plt.stem(n, s2)
plt.title(r'$s_2(n) = \cos(\frac{3\pi}{5} n)$')
plt.grid()

plt.subplot(3, 1, 3)
plt.stem(n, s3)
plt.title(r'$s_3(n) = \frac{1}{2} \sin(\frac{\pi}{2} n)$')
plt.grid()

plt.tight_layout()
```

```python

N = 100  
n = np.arange(N)  
bruit = np.random.randn(N)  

plt.figure(2)
plt.plot(n, bruit)
plt.show()

```

```python


amplitude = 2  
bruit = amplitude * np.random.randn(N)

plt.figure(2)
plt.plot(n, bruit)
plt.show()

```

```python


freq = 5  
signal_sin = np.sin(2 * np.pi * freq * n / N)  
signal_bruite = signal_sin + bruit 

plt.figure(3)
plt.plot(n, signal_bruite, label="Signal bruité")
plt.plot(n, signal_sin, linestyle="dashed", label="Signal original")
plt.show()

```

```python


def moyenne(signal):
    return np.mean(signal)  

x = np.sin(np.linspace(0, 2 * np.pi, 100))  

M = moyenne(x)
print(f"La moyenne du signal est : {M}")

def energie(signal):
    return np.sum(signal**2) 

E = energie(x)
print(f"L'énergie du signal est : {E}")
```
